```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('Webcam', frame) # Show image

    if cv2.waitKey(1) & 0xFF == ord('q'): # Checks whether q has been hit and stops the loop
        break
cap.release()
cv2.destroyAllWindows()
```
